```python
data_1_q_4 = data_1.drop(["alcohol", "quality"],axis=1)
```

```python
data_2_q_4 = data_2.drop(["alcohol", "quality"],axis=1)
```

```python
from sklearn.neighbors import DistanceMetric
```

```python
dist = DistanceMetric.get_metric('euclidean')
```

```python
pd.DataFrame(dist.pairwise(data_1_q_4.to_numpy()))
```

```python
from sklearn.metrics.pairwise import euclidean_distances,cosine_similarity
from scipy.spatial.distance import pdist
import scipy.spatial.distance
```

```python
pd.DataFrame(cosine_similarity(data_1_q_4.T), columns = data_1_q_4.columns,index = data_1_q_4.columns)
```

```python
pd.DataFrame(euclidean_distances(data_1_q_4.T), columns = data_1_q_4.columns,index = data_1_q_4.columns)
```

```python
pd.DataFrame(pdist(data_1_q_4, 'minkowski', p=1))
```

```python
pd.DataFrame(pdist(data_1_q_4, 'minkowski', p=3))
```

```python
pd.DataFrame(pdist(data_1_q_4, 'minkowski', p=5))
```

```python
pd.DataFrame(pdist(data_1_q_4, 'mahalanobis'))
```

```python
pd.DataFrame(pdist(data_1_q_4, 'correlation'))
```
